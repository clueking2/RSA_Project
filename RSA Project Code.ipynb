```python
def FME(b,n,m):
  binary_n = bin(n)
  binary_n = binary_n[2:]
  k = len(binary_n)
  x = 1
  power = b % m
  #print(power)
  for i in range(k):
    ai = int(binary_n[-i-1])
    if ai == 1:
      x = (x * power) % m
    power = (power * power) % m
  return x

FME(3,644,645)
```
